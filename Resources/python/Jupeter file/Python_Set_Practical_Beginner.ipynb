### 💡 What is a Set?

A **set** stores a collection of items — just like a list — but with two special rules:

- ✅ **No duplicates** — every item appears only once
- ✅ **No order** — items have no index (you cannot use `[0]`, `[1]`)

```python
fruits = {"Apple", "Mango", "Banana"}
```

| Feature | List | Set |
|---|---|---|
| Allows duplicates | ✅ Yes | ❌ No |
| Has index `[0]` | ✅ Yes | ❌ No |
| Uses curly braces `{}` | ❌ No | ✅ Yes |

## 🔹 Practical 4: Union and Intersection

**Scenario:** Two classes have different students. You want to combine both lists and also find which students are in both classes.

**You will learn:**
- `.union()` — combines two sets (all items, no duplicates)
- `.intersection()` — finds items that appear in **both** sets

```python
A = {1, 2, 3}
B = {3, 4, 5}

A.union(B)         # {1, 2, 3, 4, 5}   all items combined
A.intersection(B)  # {3}               only the common item
```

````python
print("=== PRACTICAL 4: Union and Intersection ===")
print()

# Two classes with some shared students (Noor and Hamza are in both!)
class_a = {"Ali", "Sara", "Noor", "Hamza"}
class_b = {"Noor", "Hamza", "Zara", "Bilal"}

print("--- Class A ---")
print(class_a)
print()

print("--- Class B ---")
print(class_b)
print()

# Step 1: Union — Combine all students from both classes, without duplicates!
# 👨‍🏫 You will type: all_students = class_a.union(class_b)


print("--- Union: All Students Combined ---")
# Print the all_students set here
# Print the total number of unique students using len()

print()

# Step 2: Intersection — Find only the students who appear in BOTH classes
# 👨‍🏫 You will type: common = class_a.intersection(class_b)


print("--- Intersection: Students in Both Classes ---")
# Print the common set here

print()

print("✅ Practical 4 Complete!")
````

## 🔹 Practical 5: Difference and Clearing a Set

**Scenario:** You want to find which students are ONLY in one class and not the other. You also want to know how to empty a set completely.

**You will learn:**
- `.difference()` — items in Set A that are **not** in Set B
- `.clear()` — removes **all** items from the set

```python

B = {3, 4, 5}

A.difference(B)  # {1, 2}   in A but NOT in B
B.difference(A)  # {4, 5}   in B but NOT in A
```

````python
print("=== PRACTICAL 5: Difference and Clearing a Set ===")
print()

class_a = {"Ali", "Sara", "Noor", "Hamza"}
class_b = {"Noor", "Hamza", "Zara", "Bilal"}

print("--- Class A ---")
print(class_a)
print()

print("--- Class B ---")
print(class_b)
print()

# Step 1: Difference — who is ONLY in Class A? (Subtract Class B from Class A)
# 👨‍🏫 You will type: only_in_a = class_a.difference(class_b)


print("--- Only in Class A (not in B) ---")
# Print only_in_a here

print()

# Step 2: Difference — who is ONLY in Class B? (Subtract Class A from Class B)
# 👨‍🏫 You will type: only_in_b = class_b.difference(class_a)


print("--- Only in Class B (not in A) ---")
# Print only_in_b here

print()

# Step 3: Clear a set — removes everything inside it, leaving an empty set
temp_set = {"X", "Y", "Z"}

print("--- Before Clear ---")
print(temp_set)
print()

# 👨‍🏫 You will type: temp_set.clear()


print("--- After Clear ---")
# Print temp_set here
# Print the len(temp_set) to prove it is empty (Count should be 0)

print()

print("✅ Practical 5 Complete!")
````
