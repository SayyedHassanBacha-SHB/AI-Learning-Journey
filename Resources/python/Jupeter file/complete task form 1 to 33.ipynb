```python
# Get input from user
num1 = float(input("Enter first number: "))
num2 = float(input("Enter second number: "))

# Calculate sum, difference, and product
sum_result = num1 + num2
difference = num1 - num2
product = num1 * num2

# Print results
print(f"\nResults:")
print(f"Sum: {num1} + {num2} = {sum_result}")
print(f"Difference: {num1} - {num2} = {difference}")
print(f"Product: {num1} × {num2} = {product}")

# Handle division - check if second number is zero
if num2 != 0:
    quotient = num1 / num2
    print(f"Quotient: {num1} ÷ {num2} = {quotient}")
else:
    # Approach: Check divisor before performing division
    # This prevents ZeroDivisionError and provides user-friendly message
    print("Quotient: Division by zero is not possible (mathematically undefined)")
```

```python
# Create two sets
set_a = {1, 2, 3, 4, 5}
set_b = {4, 5, 6, 7, 8}

# Create another set for subset check
set_c = {2, 4}

print(f"Set A: {set_a}")
print(f"Set B: {set_b}")
print(f"Set C: {set_c}")
print()

# 1. Union
union = set_a | set_b
print(f"Union (A ∪ B): {union}")

# 2. Intersection
intersection = set_a & set_b
print(f"Intersection (A ∩ B): {intersection}")

# 3. Difference (set_a - set_b)
difference = set_a - set_b
print(f"Difference (A - B): {difference}")

# 4. Symmetric Difference
sym_diff = set_a ^ set_b
print(f"Symmetric Difference (A △ B): {sym_diff}")

# 5. Check if set_a is a subset of set_c
is_subset = set_a <= set_c
print(f"\nIs Set A a subset of Set C? {is_subset}")

# Also check if set_c is a subset of set_a
is_subset2 = set_c <= set_a
print(f"Is Set C a subset of Set A? {is_subset2}")
```

```python
# Task 27: Number Classifier (Positive/Negative & Even/Odd)

numbers = [14, -7, 0, -4, 9]

for n in numbers:
    # Nested Ternary Logic
    label = "Zero" if n == 0 else (
        ("Positive Even" if n % 2 == 0 else "Positive Odd") if n > 0 else 
        ("Negative Even" if n % 2 == 0 else "Negative Odd")
    )
    
    print(f"Number: {n} -> Classification: {label}")

```
