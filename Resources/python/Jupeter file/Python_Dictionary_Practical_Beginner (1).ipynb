### 💡 What is a Dictionary?

A dictionary stores information in **key: value** pairs.

```python
person = {
    "name": "Ali",
    "age": 20
}
```

- `"name"` is the **key**
- `"Ali"` is the **value**
- You use the **key** to get the **value**: `person["name"]` → `Ali`

## 🔹 Practical 2: Update and Add Keys

**Scenario:** The student's grade has changed and we also want to add their city.

**You will learn:**
- How to **update** an existing value (same key, new value)
- How to **add** a brand new key to the dictionary

````python
print("=== PRACTICAL 2: Update and Add Keys ===")
print()

# Start with a student dictionary
student = {
    "name": "Ahmed",
    "age": 20,
    "grade": "B"
}

print("--- Original Dictionary ---")


# Step 1: Update an existing value
# 'grade' already exists — we just give it a new value
student["age"] = 25
print(student)
print("--- After Updating Grade to A ---")

# # Step 2: Add a new key
# 'city' did not exist before — this creates it
# print("--- After Adding City ---")
student["City"] = "Swat"
print(student)

# # Step 3: Add another new key

# print("--- After Adding Email ---")
student["Email"] = "icp@gmail.com"
print(student)
````

## 🔹 Practical 4: List Inside a Dictionary

**Scenario:** A student can study more than one subject. We store all subjects in a **list** as one of the dictionary values.

**You will learn:**
- How to store a list as a dictionary value
- How to access individual items inside that list
- How to add and remove items from the list

```python
# Accessing list inside dict:
student["subjects"]      # the whole list
student["subjects"][0]   # first item
student["subjects"][1]   # second item
```

````python
print("=== PRACTICAL 4: List Inside a Dictionary & Calculations ===")
print()

# Dictionary with a list stored as a value
student = {
    "name": "Sara",
    "age": 21,
    "subjects": ["Math", "Physics", "English"]
}

print("--- Original Dictionary ---")
print(student)
print()

# Step 1: Access the whole subjects list

# Step 2: Access one subject by index
print("First Subject:",)
print(student["subjects"])
print("--- Second Subject (index 1) ---")

# Add a new subject to the list at a specific position
# insert(ind, "")
student["subjects"].insert(2,"PakStudy")
# Step 3: Add a new subject to the list #append
print(student)
print("--- After Adding Chemistry & Computer Science ---")
print(student["subjects"])
# Step 4: Remove a subject from the list
student["subjects"].remove("Math")
print("--- After Removing English ---")
print(student)
# Step 5: Add a new list
# Let's add a list of marks for Sara's subjects
student["Marks"] = [22,33,55]
# print("--- After Adding Marks List ---")
print(student)
# Step 6: Do a calculation!
# Let's calculate the total of her marks using sum() and save it in the dictionary
student["Marks"] = sum(student["Marks"]) 
# print("--- After Calculating Total Score ---")
print(student["Marks"])
# Step 7: Show the full dictionary at the end
# print("--- Final Dictionary ---")
print(student)
````

### 📌 Quick Reference Card

| What you want to do | How to write it |
|---|---|
| Create a dictionary | `d = {"key": "value"}` |
| Print the whole dictionary | `print(d)` |
| Access a value | `d["key"]` |
| Update a value | `d["key"] = new_value` |
| Add a new key | `d["new_key"] = value` |
| Delete a key | `del d["key"]` |
| Access a list inside a dict | `d["list_key"][0]` |
| Add item to a list in a dict | `d["list_key"].append(item)` |
| Remove item from list in dict | `d["list_key"].remove(item)` |
| Access a dict inside a list | `my_list[0]["key"]` |
| Count items in a list | `len(my_list)` |
