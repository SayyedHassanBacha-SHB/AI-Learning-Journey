# 🔁 SECTION 1: FOR LOOPS

> A **for loop** repeats code a fixed number of times.  
> It is perfect for going through lists, datasets, or ranges of numbers.  
> In AI, for loops are used to process every sample in a dataset.

### ✅ Practice 1 — Print Numbers 1 to 10 `[Beginner]`

**Description:**  
The simplest for loop. We use `range(1, 11)` which gives us numbers from 1 to 10.  
The loop variable `i` takes each number one by one and we print it.  
This is how AI models loop through data samples one by one.

**What you learn:** Basic for loop syntax with `range()`

```python
for i in range(1, 11):
    print(i)
```

### ✅ Practice 2 — Loop Through a List of Names `[Beginner]`

**Description:**  
Instead of numbers, we can loop through a **list** — a collection of items.  
In AI, you often have a list of filenames, labels, or student records and need to process each one.  
Here we greet each student by name.

**What you learn:** Looping through a list of strings

```python
students = ["Ali", "Sara", "Umar"]

for name in students:
    print("Hello,", name)
```

### ✅ Practice 3 — Sum of All Numbers in a List `[Beginner]`

**Description:**  
We create a variable `total = 0` before the loop.  
Each time the loop runs, we **add** the current number to total.  
This is exactly how an AI computes the total loss or total score across all training examples.

**What you learn:** Accumulating values inside a loop

```python
numbers = [10, 20, 30, 40]
total = 0

for n in numbers:
    total = total + n

print("Total:", total)
```

### ✅ Practice 4 — Find the Average of Scores `[Intermediate]`

**Description:**  
After summing, we divide by the count to find the average.  
We use `len(scores)` to get how many items are in the list.  
Calculating averages is fundamental in AI — average accuracy, average loss per epoch, etc.

**What you learn:** Using `len()` and computing averages

```python
scores = [85, 90, 78, 92, 88]
total = 0

for score in scores:
    total = total + score

average = total / len(scores)
print("Average Score:", average)
```

### ✅ Practice 5 — Multiplication Table `[Intermediate]`

**Description:**  
We use **f-strings** (formatted strings) to make output neat.  
`f"..."` lets us put variables directly inside a string using `{}`.  
This teaches clean printing — useful when you display model evaluation results.

**What you learn:** f-strings and neat output formatting

```python
num = 5

for i in range(1, 11):
    print(f"{num} x {i} = {num * i}")
```

### ✅ Practice 6 — Counting Even and Odd Numbers `[Intermediate]`

**Description:**  
We combine a for loop with an **if-else** inside it.  
The **% operator** (modulus) gives the remainder of division.  
If the remainder when divided by 2 is 0, the number is even.  
This pattern is used to filter data in AI preprocessing.

**What you learn:** Combining loops with conditions (modulus operator)

```python
numbers = [1, 2, 3, 4, 5, 6]
even_count = 0
odd_count = 0

for n in numbers:
    if n % 2 == 0:
        even_count += 1
    else:
        odd_count += 1

print("Even:", even_count, "| Odd:", odd_count)
```

### ✅ Practice 7 — Nested Loop: Simple Pattern `[Advanced]`

**Description:**  
A **nested loop** is a loop inside a loop.  
The outer loop runs 3 times (rows), and for each row the inner loop runs 3 times (columns).  
Nested loops are used in AI to process 2D data like images (rows and columns of pixels).

**What you learn:** Nested loops — loop inside a loop

```python
for row in range(1, 4):
    for col in range(1, 4):
        print(f"Row{row}-Col{col}", end="  ")
    print()  # new line after each row
```

### ✅ Practice 8 — List Comprehension (Short Loop) `[Advanced]`

**Description:**  
Python allows a **shorter way** to write a for loop that creates a list — called **list comprehension**.  
Instead of 4 lines, you write 1.  
AI code uses this all the time to quickly transform datasets — like squaring all values or filtering samples.

**What you learn:** List comprehension — compact loop syntax

```python
# Normal way (4 lines)
squares = []
for x in range(1, 6):
    squares.append(x * x)
print("Normal way:", squares)

# Short way — list comprehension (1 line)
squares = [x * x for x in range(1, 6)]
print("List comprehension:", squares)
```

### ✅ Practice 9 — Normalize Data (AI Preprocessing) `[Expert]`

**Description:**  
Normalization scales all values to be between 0 and 1.  
This is one of the most important steps in real AI models.  
We find the minimum and maximum first, then for each value apply the formula:  
`(value - min) / (max - min)`

**What you learn:** Real-world AI data normalization

```python
data = [10, 20, 30, 40, 50]
min_val = min(data)
max_val = max(data)

normalized = []
for val in data:
    n = (val - min_val) / (max_val - min_val)
    normalized.append(n)

print("Original: ", data)
print("Normalized:", normalized)
```

### ✅ Practice 10 — Simulate Training Epochs `[Expert]`

**Description:**  
In AI, we train a model for many **epochs** (full passes over the data).  
This simulates that — each epoch, the loss decreases a little.  
We track the best epoch (lowest loss).  
This pattern appears in every real AI training script like PyTorch or Keras.

**What you learn:** Simulating AI training loop logic with `enumerate()`

```python
losses = [0.9, 0.7, 0.5, 0.4, 0.35, 0.3]
best_epoch = 0
best_loss = losses[0]

for epoch, loss in enumerate(losses):
    print(f"Epoch {epoch+1}: Loss = {loss}")
    if loss < best_loss:
        best_loss = loss
        best_epoch = epoch + 1

print(f"\n🏆 Best Epoch: {best_epoch} with Loss {best_loss}")
```

# 🔄 SECTION 2: WHILE LOOPS

> A **while loop** keeps running as long as a condition is True.  
> Use it when you **don't know** how many times you need to repeat.  
> In AI, while loops are used for training until convergence, retrying until success, etc.

### ✅ Practice 1 — Count from 1 to 5 `[Beginner]`

**Description:**  
We start with `count = 1`. The while loop checks: "is count ≤ 5?"  
If yes, it prints and adds 1 to count.  
When count reaches 6, the condition becomes False and the loop stops.  
⚠️ **Always remember to increase the counter or it will loop forever!**

**What you learn:** Basic while loop with a counter

```python
count = 1

while count <= 5:
    print(count)
    count = count + 1
```

### ✅ Practice 2 — Ask User Until Correct Answer `[Beginner]`

**Description:**  
While loops shine when you **don't know how many tries** it will take.  
Here we keep asking until the user types the right answer.  
This pattern is used in chatbots that keep asking until they understand the user's intent.

**What you learn:** Loop until a condition is met by user input

```python
answer = ""

while answer != "python":
    answer = input("What language is used in AI? ").lower()

print("Correct! 🎉")
```

### ✅ Practice 3 — Countdown Timer `[Beginner]`

**Description:**  
We count **down** instead of up by subtracting 1 each time.  
When the count hits 0, we print "Blast off!".  
Countdown logic is used in scheduled AI tasks — like running a job after N seconds.

**What you learn:** Decrementing counter in a while loop

```python
count = 5

while count > 0:
    print(f"{count}...")
    count -= 1

print("Blast off! 🚀")
```

### ✅ Practice 4 — Sum Numbers Until User Stops `[Intermediate]`

**Description:**  
The user keeps entering numbers and we keep adding them.  
When they type **-1** (the "stop signal"), the loop ends.  
This is called a **sentinel value** — a special value that signals the end of input.  
Common in data entry programs and streaming data pipelines.

**What you learn:** Sentinel value to stop a while loop

```python
total = 0
num = 0

while num != -1:
    num = int(input("Enter a number (-1 to stop): "))
    if num != -1:
        total += num

print("Total:", total)
```

### ✅ Practice 5 — Password Attempt Limit `[Intermediate]`

**Description:**  
We give only 3 attempts. Each wrong guess increases the counter.  
If the user gets it right, we use **break** to exit immediately.  
If they fail all 3 times, we lock them out.  
This models real login security systems used in AI applications.

**What you learn:** Using `break` inside a while loop

```python
correct = "ai2024"
attempts = 0

while attempts < 3:
    pwd = input("Enter password: ")
    if pwd == correct:
        print("Access Granted ✅")
        break
    attempts += 1
    print(f"Wrong! {3 - attempts} tries left")
else:
    print("Locked Out ❌")
```

### ✅ Practice 6 — Find First Number Divisible by 7 and 5 `[Intermediate]`

**Description:**  
We search from 1 upward until we find a number that satisfies **both conditions** at once.  
The **and** keyword means both must be True.  
Searching with conditions is essential in AI data filtering and rule engines.

**What you learn:** Using `and` / `or` conditions inside while + `break`

```python
num = 1

while True:
    if num % 7 == 0 and num % 5 == 0:
        print(f"Found it: {num}")
        break
    num += 1
```

### ✅ Practice 7 — Simulate a Simple Chatbot `[Advanced]`

**Description:**  
A simple chatbot keeps running in a **while True** loop (runs forever until stopped).

Typing "bye" exits the loop.

### ✅ Practice 8 — Fibonacci Sequence `[Advanced]`

**Description:**  
The Fibonacci sequence is where each number is the sum of the two before it: 0, 1, 1, 2, 3, 5, 8...  
It appears in nature, finance, and AI research.  
We use two variables and keep swapping them — a classic while loop pattern.

### ✅ Practice 9 — Gradient Descent Simulation `[Expert]`

**Description:**  
**Gradient Descent** is the core algorithm that trains AI models!  
We start at a position and keep moving toward the minimum by subtracting `learning_rate × gradient`.  
We stop when the value is very close to 0 (found the minimum).  
This is exactly how neural networks learn during training.

**What you learn:** Real AI optimization algorithm with a while loop

```python
# Minimize f(x) = x^2  -->  minimum is at x = 0
x = 10.0
lr = 0.1    # learning rate
step = 0

while abs(x) > 0.01:
    gradient = 2 * x      # derivative of x^2 is 2x
    x = x - lr * gradient
    step += 1
    print(f"Step {step}: x = {x:.4f}")

print(f"\n✅ Minimum found at x ≈ {x:.4f} in {step} steps")
```

### ✅ Practice 10 — Binary Search Algorithm `[Expert]`

**Description:**  
Binary search is a fast way to find an item in a sorted list.  
Instead of checking every item, we check the **middle**, then go left or right.  
This halves the search space each time — much faster than checking one by one.  
Used in AI for hyperparameter tuning and decision tree lookups.

**What you learn:** Binary search — one of the most important algorithms in CS & AI

```python
data = [1, 5, 10, 20, 35, 50, 80]
target = 35
left, right = 0, len(data) - 1

while left <= right:
    mid = (left + right) // 2
    print(f"Checking index {mid} → value {data[mid]}")

    if data[mid] == target:
        print(f"\n✅ Found {target} at index {mid}")
        break
    elif data[mid] < target:
        left = mid + 1   # go right
    else:
        right = mid - 1  # go left
```

### ✅ Practice 8 — Spam Detector (Keyword Check) `[Advanced]`

**Description:**

If even one matches, we flag it as spam.  
This is a simplified rule-based spam filter — before ML, this is how ALL spam filters worked!

### ✅ Practice 9 — Simple Classifier (Threshold) `[Expert]`

**Description:**  
In AI, a model outputs a **probability** (a number between 0 and 1).  
We use if-else to turn that number into a **class label**.  
If probability ≥ 0.5 → Positive, else → Negative.  
This step is called **thresholding** and is used in every binary classification model.

**What you learn:** Converting probability scores to class labels

```python
# Simulated model output probabilities
predictions = [0.85, 0.3, 0.67, 0.1, 0.92]
threshold = 0.5

for prob in predictions:
    if prob >= threshold:
        label = "Positive 😊"
    else:
        label = "Negative 😟"
    print(f"Probability={prob:.2f} → {label}")
```

### ✅ Practice 10 — Model Accuracy Evaluator `[Expert]`

**Description:**  
We compare the AI model's **predicted labels** vs the **true labels** and count how many match.  
Then we calculate accuracy as a percentage.  
This is literally how you evaluate every AI model — comparing predictions to ground truth.

**What you learn:** Model evaluation using if statements — the foundation of AI metrics

```python
true_labels = [1, 0, 1, 1, 0, 1]
predicted   = [1, 0, 0, 1, 0, 1]
correct = 0

for i in range(len(true_labels)):
    if true_labels[i] == predicted[i]:
        correct += 1

accuracy = (correct / len(true_labels)) * 100
print(f"Correct Predictions: {correct}/{len(true_labels)}")
print(f"Model Accuracy: {accuracy:.2f}%")
```

### ✅ Practice 5 — Function Returning Multiple Values `[Intermediate]`

**Description:**  
Python functions can return **more than one value** using a comma.  
The values come back as a **tuple**, which we can unpack into separate variables.  
In AI, a function might return both the prediction and the confidence score, or both min and max of a dataset.

**What you learn:** Returning multiple values (tuple unpacking)

```python
def get_stats(numbers):
    minimum = min(numbers)
    maximum = max(numbers)
    average = sum(numbers) / len(numbers)
    return minimum, maximum, average

low, high, avg = get_stats([10, 30, 20, 40])
print(f"Min={low}, Max={high}, Average={avg}")
```

### ✅ Practice 6 — Recursive Function: Factorial `[Intermediate]`

**Description:**  
A **recursive function** calls itself!  
Factorial of 5 = 5 × 4 × 3 × 2 × 1 = 120.  
The function keeps calling itself with a smaller number until it reaches 1 (the **base case**).  
Recursion is used in AI tree search algorithms and many mathematical computations.

**What you learn:** Recursion — a function that calls itself

```python
def factorial(n):
    if n == 1:              # base case — stop recursing here!
        return 1
    return n * factorial(n - 1)   # call itself with smaller n

print(factorial(5))   # 5*4*3*2*1 = 120
print(factorial(7))   # 5040
```

### ✅ Practice 7 — Lambda: Anonymous Functions `[Advanced]`

**Description:**  
A **lambda** is a tiny one-line function without a name.  
Syntax: `lambda parameters: expression`  
Lambda functions are everywhere in data science — used with `sorted()`, `map()`, `filter()`, and pandas DataFrames.

**What you learn:** Lambda (anonymous) functions

```python
# Regular function
def double(x):
    return x * 2

# Lambda version — same thing in one line!
double = lambda x: x * 2

print(double(5))    # 10
print(double(9))    # 18

# Sorting a list of tuples by the second item using lambda
pairs = [(1, "B"), (3, "A"), (2, "C")]
pairs.sort(key=lambda x: x[1])
print(pairs)   # [(3,'A'), (1,'B'), (2,'C')]
```

### ✅ Practice 8 — Normalize Data Function `[Advanced]`

**Description:**  
We now wrap the normalization logic into a **reusable function**.  
This is software engineering best practice — write code once, use it many times.  
In real AI projects, you'd call this function on every new dataset you receive.

**What you learn:** Wrapping logic into a reusable AI utility function

```python
def normalize(data):
    min_val = min(data)
    max_val = max(data)
    return [(x - min_val) / (max_val - min_val) for x in data]

dataset1 = [10, 20, 30, 40]
dataset2 = [5, 50, 100]

print("Dataset 1 normalized:", normalize(dataset1))
print("Dataset 2 normalized:", normalize(dataset2))
```

### ✅ Practice 9 — Higher-Order Functions: map() and filter() `[Expert]`

**Description:**  
**Higher-order functions** take other functions as input.  
- `map()` applies a function to every item in a list  
- `filter()` keeps only items where the function returns True  
These are core functional programming tools heavily used in AI data pipelines.

**What you learn:** `map()` and `filter()` — functional programming in AI

```python
numbers = [1, 2, 3, 4, 5, 6]

# map: apply a function to every item
squared = list(map(lambda x: x**2, numbers))
print("Squared:", squared)   # [1, 4, 9, 16, 25, 36]

# filter: keep only items that pass the test
evens = list(filter(lambda x: x % 2 == 0, numbers))
print("Evens:", evens)       # [2, 4, 6]

# Combine: square only the even numbers
result = list(map(lambda x: x**2, filter(lambda x: x % 2 == 0, numbers)))
print("Even squares:", result)  # [4, 16, 36]
```

### ✅ Practice 10 — Build a Mini Neural Network Neuron `[Expert]`

**Description:**  
We combine everything — functions, loops, conditionals — to simulate one neuron of a neural network.  
The neuron takes **inputs + weights**, computes the weighted sum, then applies an **activation function** (sigmoid) to squash the output between 0 and 1.  
**This is the foundation of deep learning!** Every layer in GPT, BERT, or any neural network is made of neurons like this.

**What you learn:** Simulating a real neuron — foundation of neural networks

```python
import math

# Sigmoid squashes any number to between 0 and 1
def sigmoid(x):
    return 1 / (1 + math.exp(-x))

# A single neuron: takes inputs and weights, returns a prediction
def neuron(inputs, weights, bias):
    weighted_sum = 0
    for i in range(len(inputs)):
        weighted_sum += inputs[i] * weights[i]
    weighted_sum += bias         # add bias
    return sigmoid(weighted_sum) # apply activation

# Test the neuron
inputs  = [0.5, 0.8, 0.2]
weights = [0.4, 0.6, 0.9]
bias    = -0.3

output = neuron(inputs, weights, bias)
print(f"Neuron Output: {output:.4f}")
print(f"Interpretation: {'Positive (>0.5)' if output > 0.5 else 'Negative (<0.5)'}")
```
