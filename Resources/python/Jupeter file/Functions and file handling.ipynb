### Example 4 — Intermediate
A function that calculates a student's total marks, average, and grade.

```python
def student_result(name, maths, english, science):
    total = maths + english + science
    average = total / 3
    if average >= 80:
        grade = "A"
    elif average >= 60:
        grade = "B"
    else:
        grade = "C"
    print("Student:", name)
    print("Total Marks:", total)
    print("Average:", round(average, 2))
    print("Grade:", grade)
    print("-" * 30)

student_result("Ahmed", 85, 78, 90)
student_result("Zara", 55, 62, 70)
student_result("Ali" , 66,77,55)
```

### Example 5 — Upper-Intermediate
A shopping bill calculator with item name, price, quantity, and optional discount.

```python
def calculate_bill(item, price, quantity, discount_percent=0):
    subtotal = price * quantity
    discount_amount = subtotal * (discount_percent / 100)
    final_bill = subtotal - discount_amount
    print("Item      :", item)
    print("Unit Price:", price)
    print("Quantity  :", quantity)
    print("Subtotal  :", subtotal)
    print("Discount  :", discount_amount)
    print("Final Bill:", final_bill)
    print("-" * 30)
    return final_bill

calculate_bill("Rice (kg)", 120, 3)
calculate_bill("Shirt", 800, 2, 10)
calculate_bill("Laptop", 50000, 1, 5)
```

## Topic 2: Functions and Variable Scope
---

### Example 1 — Very Basic
A local variable exists only inside the function where it is created.

```python
def show_city():
    city = "Lahore"
    print("Inside function, city is:", city)

show_city()
```

### Example 2 — Basic
A global variable is defined outside the function and can be read inside it.

```python
school_name = "City School"

def show_school():
    print("School Name:", school_name)

show_school()
print("Also accessible outside:", school_name)
```

### Example 3 — Basic-Intermediate
Local variable has priority over a global variable with the same name inside a function.

```python
temperature = 30
def check_weather():
    # temperature = 2
    print("Inside function, temperature:", temperature)

check_weather()
print("Outside function, temperature:", temperature)
```

```python
def add_with_print(a, b):
    result = a + b
    print(result)        # shows on screen but gives NOTHING back

def add_with_return(a, b):
    result = a + b
    return result        # gives the value BACK to caller
x = add_with_print(5, 3)
print("x is:", x)
y = add_with_return(5, 3)
print("y is:", y)
```

```python
def calculate_tax(price):
    tax = price * 0.17
    return tax

def calculate_total(price):
    t = calculate_tax(price)     # return value caught here
    total = price + t
    return total                  # passed up again

bill = calculate_total(1000)
print(bill)
```

```python
def check_number(n):
    if n > 0:
        return "Positive"
    print("This line only runs if n is NOT positive")
    return "Zero or Negative"

print(check_number(5))
print(check_number(-3))
```

### Example 4 — Intermediate
Using the `global` keyword to modify a global variable from inside a function.

```python
attendance_count = 0

def mark_attendance():
    global attendance_count
    attendance_count += 1
    print("Attendance marked. Total present:", attendance_count)

mark_attendance()
mark_attendance()
mark_attendance()
print("Final attendance count:", attendance_count)
```

```python
current_level = 1

def level_up():
    global current_level
    current_level += 1
    print(f"CONGRATULATIONS! You reached Level {current_level}")

# The player plays the game...
level_up()  # Reaches Level 2
level_up()  # Reaches Level 3
level_up()
print("Game Over. Your final level was:", current_level)
```

```python
wrong_attempts = 0

def enter_wrong_password():
    global wrong_attempts
    wrong_attempts += 1
    
    if wrong_attempts >= 3:
        print("!!! CARD BLOCKED !!! Please contact the bank.")
    else:
        print(f"Wrong password. You have {3 - wrong_attempts} tries left.")

# User tries 3 times
enter_wrong_password()
enter_wrong_password()
enter_wrong_password()
```

### Example 5 — Upper-Intermediate
A bank balance example that shows the difference between local changes and global state.

```python
bank_balance = 10000

def deposit(amount):
    global bank_balance
    bank_balance += amount
    print("Deposited:", amount, "| New Balance:", bank_balance)

def withdraw(amount):
    global bank_balance
    if amount > bank_balance:
        print("Insufficient balance! Current:", bank_balance)
    else:
        bank_balance -= amount
        print("Withdrawn:", amount, "| New Balance:", bank_balance)

def show_balance():
    print("Current Balance:", bank_balance)

show_balance()
deposit(5000)
withdraw(3000)
withdraw(15000)
show_balance()
```

```python
def get_stats(numbers):
    minimum = min(numbers)
    maximum = max(numbers)
    average = sum(numbers) / len(numbers)
    return minimum, maximum, average

lo, hi, avg = get_stats([10, 20, 30, 40, 50])
print("Min:", lo)
print("Max:", hi)
print("Avg:", avg)
```

```python
def get_grade(marks):
    if marks >= 80:
        return "A"          # returns a string

def get_passed_students(all_marks):
    passed = []
    for m in all_marks:
        if m >= 50:
            passed.append(m)
    return passed           # returns a list

def is_adult(age):
    return age >= 18        # returns a boolean (True or False)

print(get_grade(85))                           # A
print(get_passed_students([45, 78, 32, 91]))   # [78, 91]
print(is_adult(20))                            # True
print(is_adult(15))                            # False
```

## Topic 3: Lambda Expressions
---

### Example 1 — Very Basic
A lambda that doubles a number — like doubling the price of an item.

```python
double = lambda x: x * 2
print(double(5))
print(double(50))
```

### Example 2 — Basic
A lambda that converts Celsius to Fahrenheit temperature.

```python
to_fahrenheit = lambda c: (c * 9 / 5) + 32
print("0 C  =", to_fahrenheit(0), "F")
print("25 C =", to_fahrenheit(37.5), "F")
print("100 C=", to_fahrenheit(100), "F")
```

### Example 3 — Basic-Intermediate
A lambda that calculates the discounted price of a product.

```python
discounted_price = lambda price, discount: price - (price * discount / 100)
print("Shirt after 10% off:", discounted_price(500, 10))
print("Phone after 15% off:", discounted_price(30000, 15))
print("Book after 5% off  :", discounted_price(200, 5))
```

### Example 4 — Intermediate
Using lambda with `sorted()` to sort a list of students by their marks.

```python
students = [
    ("Ali", 78),
    ("Sara", 92),
    ("John", 65)
]

sorted_students = sorted(students, key=lambda x: x[1], reverse=True)

print("Students sorted by marks:")

for student in sorted_students:
    print(student)
```

### Example 5 — Upper-Intermediate
Using multiple lambdas to build a simple shopping bill processor.

```python
apply_tax = lambda amount: amount + amount * 0.08
calc_total = lambda price, qty: price * qty

item_name = "Rice"
price = 100
quantity = 2

total = calc_total(price, quantity)
final_price = apply_tax(total)

print("Item:", item_name)
print("Total before tax:", total)
print("Total after 8% tax:", final_price)
```

## Topic 4: Map and Filter
---

### Example 1 — Very Basic
Using `map()` to convert a list of temperatures from Celsius to Fahrenheit.

```python
celsius_temps = [0, 20, 37, 100]
fahrenheit_temps = list(map(lambda c: (c * 9/5) + 32, celsius_temps))
print("Celsius   :", celsius_temps)
print("Fahrenheit:", fahrenheit_temps)
```

### Example 2 — Basic
Using `filter()` to keep only students who passed (marks >= 50).

```python
marks = [45, 78, 32, 90, 55, 48, 88, 61]
passed = list(filter(lambda m: m >= 50, marks))
failed = list(filter(lambda m: m < 50, marks))
print("All Marks:", marks)
print("Passed   :", passed)
print("Failed   :", failed)
```

### Example 3 — Basic-Intermediate
Using `map()` to apply a 10% price increase to a shopping list.

```python
prices = [100, 250, 430, 75, 990]
new_prices = list(map(lambda p: round(p * 1.10, 2), prices))
print("Original Prices :", prices)
print("Prices After 10%:", new_prices)
```

### Example 4 — Intermediate
Using both `map()` and `filter()` to find adults from a list of ages and calculate discounted ticket prices.

```python
ages = [12, 25, 8, 34, 16, 42, 19, 6]
adults = list(filter(lambda age: age >= 18, ages))
print("All Ages :", ages)
print("Adults   :", adults)

ticket_price = 500
adult_prices = list(map(lambda age: ticket_price, adults))
print("Ticket prices for adults:", adult_prices)
print("Total ticket revenue:", sum(adult_prices))
```

### Example 5 — Upper-Intermediate
A student report system: filter passed students, map their letter grades, and summarise.

```python
students = [
    {"name": "Ali",   "marks": 88},
    {"name": "Sara",  "marks": 45},
    {"name": "John",  "marks": 72},
    {"name": "Mia",   "marks": 35},
    {"name": "Ahmed", "marks": 91},
    {"name": "Hina",  "marks": 60},
]

passed_students = list(filter(lambda s: s["marks"] >= 50, students))

get_grade = lambda m: "A" if m >= 80 else ("B" if m >= 65 else "C")
with_grades = list(map(lambda s: {"name": s["name"], "marks": s["marks"], "grade": get_grade(s["marks"])}, passed_students))

print("Passed Students with Grades:")
for s in with_grades:
    print(f"  {s['name']:<8} Marks: {s['marks']}  Grade: {s['grade']}")
```

## Topic 5: Inner / Nested Functions
---

### Example 1 — Very Basic
A function inside another function — a simple greeting with a title added.

```python
def greet_with_title(name):
    def add_title():
        return "Mr./Ms. " + name
    full_name = add_title()
    print("Welcome,", full_name)

greet_with_title("Ali")
greet_with_title("Sara")
```

### Example 2 — Basic
An outer function for processing a student mark, with an inner function deciding pass or fail.

```python
def process_mark(name, marks):
    def get_result():
        if marks >= 50:
            return "PASS"
        else:
            return "FAIL"
    result = get_result()
    print(name, "scored", marks, "—", result)

process_mark("Ahmed", 72)
process_mark("Zara", 43)
```

### Example 3 — Basic-Intermediate
A shopping bill function with inner helpers for tax and discount calculations.

```python
def generate_bill(item, price, quantity):
    def calculate_subtotal():
        return price * quantity

    def apply_tax(subtotal):
        return subtotal * 1.17

    subtotal = calculate_subtotal()
    total    = apply_tax(subtotal)
    print("Item     :", item)
    print("Subtotal :", subtotal)
    print("Total+Tax:", round(total, 2))
    print("-" * 25)

generate_bill("Tea Box", 150, 3)
generate_bill("Headphones", 2500, 2)
```

### Example 4 — Intermediate
A password validator with separate inner functions for length, uppercase, and digit checks.

```python
def validate_password(password):
    def is_long_enough():
        return len(password) >= 8

    def has_uppercase():
        return any(c.isupper() for c in password)

    def has_digit():
        return any(c.isdigit() for c in password)

    if not is_long_enough():
        print(password, "-> Too short (min 8 characters)")
    elif not has_uppercase():
        print(password, "-> Must have at least one uppercase letter")
    elif not has_digit():
        print(password, "-> Must have at least one digit")
    else:
        print(password, "-> Password is STRONG")

validate_password("abc")
validate_password("abcdefgh")
validate_password("Abcdefgh")
validate_password("Abcdefg9")
```

### Example 5 — Upper-Intermediate
A student report card system using nested functions for formatting, grading, and printing.

```python
def generate_report_card(name, subjects_marks):
    def calculate_average():
        return sum(subjects_marks.values()) / len(subjects_marks)

    def assign_grade(avg):
        if avg >= 85:
            return "A+"
        elif avg >= 75:
            return "A"
        elif avg >= 60:
            return "B"
        elif avg >= 50:
            return "C"
        else:
            return "F"

    def print_card(avg, grade):
        print("=" * 35)
        print("  REPORT CARD")
        print("  Student:", name)
        print("-" * 35)
        for subject, marks in subjects_marks.items():
            print(f"  {subject:<15}: {marks}")
        print("-" * 35)
        print("  Average :", round(avg, 2))
        print("  Grade   :", grade)
        print("=" * 35)

    avg   = calculate_average()
    grade = assign_grade(avg)
    print_card(avg, grade)

generate_report_card("Ahmed Khan", {"Maths": 88, "English": 76, "Science": 91, "Urdu": 70})
generate_report_card("Sara Ali",   {"Maths": 55, "English": 62, "Science": 48, "Urdu": 58})
```

### Example 1 — Very Basic

### Example 2 — Basic
Save a list of student names to a file, one name per line, then display them.

```python
students = ["Ali", "Sara", "John", "Mia", "Ahmed"]

with open("students.txt", "w") as f:
    for name in students:
        f.write(name + "\n")

print("Students saved to file.")
print("\nReading back from file:")
with open("students.txt", "r") as f:
    for line in f:
        print(" -", line.strip())
```

### Example 3 — Basic-Intermediate
Append new student names to an existing file without overwriting old data.

```python
new_students = ["Hina", "Bilal", "Nadia"]

with open("students.txt", "a") as f:
    for name in new_students:
        f.write(name + "\n")

print("New students appended.")
print("\nFull list now:")
with open("students.txt", "r") as f:
    lines = f.readlines()
    print("Total students:", len(lines))
    for line in lines:
        print(" -", line.strip())
```

### Example 4 — Intermediate
Save an attendance record (name and status) to a file and read it as a report.

```python
attendance = [
    ("Ali",   "Present"),
    ("Sara",  "Absent"),
    ("John",  "Present"),
    ("Mia",   "Present"),
    ("Ahmed", "Absent"),
]

with open("attendance.txt", "w") as f:
    f.write("Name         | Status\n")
    f.write("-" * 25 + "\n")
    for name, status in attendance:
        f.write(f"{name:<13}| {status}\n")

print("Attendance saved.")
print("\n--- Attendance Report ---")
with open("attendance.txt", "r") as f:
    print(f.read())
```

### Example 5 — Upper-Intermediate
A simple banking ledger: save transactions (deposit/withdraw) to a file and generate a mini statement.

```python
transactions = [
    ("Deposit",  5000),
    ("Deposit",  3000),
    ("Withdraw", 1500),
    ("Deposit",  2000),
    ("Withdraw",  800),
]

with open("bank_ledger.txt", "w") as f:
    f.write("=== Bank Statement ===\n")
    balance = 0
    for txn_type, amount in transactions:
        if txn_type == "Deposit":
            balance += amount
        else:
            balance -= amount
        f.write(f"{txn_type:<10}: {amount:>8}   Balance: {balance}\n")
    f.write("=" * 35 + "\n")
    f.write(f"Final Balance: {balance}\n")

print("Ledger saved.\n")
with open("bank_ledger.txt", "r") as f:
    print(f.read())
```

## Topic 7: Exception Handling
---

### Example 1 — Very Basic
Handle a simple divide-by-zero error using try and except.

```python
try:
    result = 10 / 0
    print("Result:", result)
except ZeroDivisionError:
    print("Error: You cannot divide by zero!")
```

### Example 2 — Basic
Handle a wrong input type — user enters a word instead of a number for age.

```python
def check_age(age_input):
    try:
        age = int(age_input)
        if age >= 18:
            print("You are an adult.")
        else:
            print("You are a minor.")
    except ValueError:
        print("Invalid input! Please enter a number.")

check_age("25")
check_age("fifteen")
check_age("16")
```

### Example 3 — Basic-Intermediate
A calculator with multiple exception handlers for division and invalid input.

```python
def safe_divide(a, b):
    try:
        result = int(a) / int(b)
        print("Result:", result)
    except ZeroDivisionError:
        print("Error: Cannot divide by zero.")
    except ValueError:
        print("Error: Please enter valid numbers only.")

safe_divide(10, 2)
safe_divide(10, 0)
safe_divide(10, "abc")
```

### Example 4 — Intermediate
A file reader with exception handling for missing file and permission errors, plus a `finally` block.

```python
def read_student_file(filename):
    try:
        with open(filename, "r") as f:
            contents = f.read()
            print("File contents:")
            print(contents)
    except FileNotFoundError:
        print("Error: File not found —", filename)
    except PermissionError:
        print("Error: No permission to read this file.")
    finally:
        print("File operation attempted for:", filename)

read_student_file("students.txt")
read_student_file("grades_2030.txt")
```

### Example 5 — Upper-Intermediate
A complete ATM simulator with exception handling for wrong input, insufficient balance, and a transaction log.

```python
class InsufficientFundsError(Exception):
    pass

def atm_withdraw(balance, amount_input):
    try:
        amount = float(amount_input)
        if amount <= 0:
            raise ValueError("Withdrawal amount must be positive.")
        if amount > balance:
            raise InsufficientFundsError("Balance too low for this withdrawal.")
        balance -= amount
        print(f"Withdrawn: {amount}  | Remaining Balance: {balance}")
        return balance
    except ValueError as e:
        print("Input Error:", e)
        return balance
    except InsufficientFundsError as e:
        print("Transaction Denied:", e)
        return balance
    finally:
        print("ATM transaction processed.\n")

balance = 8000
balance = atm_withdraw(balance, "2000")
balance = atm_withdraw(balance, "abc")
balance = atm_withdraw(balance, "-500")
balance = atm_withdraw(balance, "9000")
```
