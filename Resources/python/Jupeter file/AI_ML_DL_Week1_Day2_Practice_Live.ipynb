## Section 1: Quick Recap

### Key Concepts from Day 2:

- **Lists `[]`**: Ordered, mutable, allow duplicates. Methods: `append()`, `insert()`, `remove()`, `pop()`, `sort()`, `reverse()`, `extend()`, `count()`, `index()`
- **Tuples `()`**: Ordered, immutable (cannot change after creation), allow duplicates. Faster than lists, used for fixed data
- **Dictionaries `{}`**: Key-value pairs, unordered (but insertion order maintained in Python 3.7+), keys must be unique. Methods: `keys()`, `values()`, `items()`, `get()`, `update()`, `pop()`
- **Sets `set()`**: Unordered, mutable, only unique items (no duplicates). Operations: `union()`, `intersection()`, `difference()`, `add()`, `remove()`
- **List Indexing**: Zero-based, positive (0, 1, 2...) and negative (-1, -2, -3...)
- **List Slicing**: `list[start:end:step]` - end is exclusive
- **List Comprehension**: `[expression for item in iterable if condition]` - elegant way to create lists
- **When to Use**: Lists for ordered modifiable data, Tuples for fixed data, Dictionaries for key-value lookups, Sets for unique items

## 📚 Practice Tips for Data Structures

### How to Choose the Right Data Structure:

**Use Lists when:**
- Order matters
- You need to modify data frequently
- You need duplicate values
- You need to access by position/index

**Use Tuples when:**
- Data should not change
- You need faster performance than lists
- Using as dictionary keys
- Returning multiple values from functions

**Use Dictionaries when:**
- You need key-value relationships
- Fast lookup by unique identifier
- Data has named attributes
- Building configurations or settings

**Use Sets when:**
- You need only unique items
- Mathematical operations (union, intersection, difference)
- Fast membership testing
- Removing duplicates from lists

### Common Mistakes to Avoid:

- Trying to modify tuples (they're immutable!)
- Using lists when sets would be more efficient
- Not checking if key exists before accessing dictionary
- Forgetting that sets are unordered
- Modifying list while iterating over it
- Using mutable objects as dictionary keys

### Performance Considerations:

- List append: O(1) - Fast
- List insert at beginning: O(n) - Slow
- Dictionary lookup: O(1) - Very fast
- Set membership test: O(1) - Very fast
- List membership test: O(n) - Slow for large lists

## 🎓 Preparing for Job Interviews

### Common Interview Questions (Day 2 Level):

1. **What's the difference between a list and a tuple?**
   - Lists are mutable, tuples are immutable
   - Tuples are faster and use less memory
   - Tuples can be dictionary keys

2. **How do you remove duplicates from a list?**
   - Convert to set: `list(set(my_list))`
   - Or use dict.fromkeys(): `list(dict.fromkeys(my_list))`

3. **What's the difference between append() and extend()?**
   - append() adds single element
   - extend() adds all elements from iterable

4. **How do you merge two dictionaries?**
   - Use update(): `dict1.update(dict2)`
   - Or unpacking: `{**dict1, **dict2}`

5. **When would you use a set instead of a list?**
   - When you need unique items only
   - When order doesn't matter
   - When you need fast membership testing
   - When you need set operations
