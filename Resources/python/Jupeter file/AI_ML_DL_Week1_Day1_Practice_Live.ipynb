## Section 1: Quick Recap

### 🏢 Exercise 1: Employee Salary Calculator

**Scenario**: You're working as a junior developer at a company. Your first task is to create a salary calculator.

**Requirements**:
1. Get employee name (string)
2. Get basic salary (float)
3. Get overtime hours (integer)
4. Calculate overtime pay: overtime_hours × 500 PKR per hour
5. Calculate total salary: basic_salary + overtime_pay
6. Calculate tax (5% of total salary)
7. Calculate net salary (after tax)
8. Display a formatted salary slip

**Expected Output Format**:
```
=== SALARY SLIP ===
Employee: Ahmed Ali
Basic Salary: PKR 50000.00
Overtime Pay: PKR 5000.00
Total Salary: PKR 55000.00
Tax (5%): PKR 2750.00
Net Salary: PKR 52250.00
===================
```

````python
# Write your code here
em_name = input("Enter employee name : ")
em_salary = float(input("Enter employee salary : "))
overtime_hours = float(input("Enter overtime hours of employee : "))
overtime_pay = overtime_hours * 500
total_salary = em_salary + overtime_pay
tax = 0.05 * total_salary
net_salary = total_salary - tax

print("=======SALARY SLIP========")
print(f"Employe         :   {em_name}")
print(f"Basic Salary    : PKR {em_salary}")
print(f"Overtime Pay    : PKR {overtime_pay}")
print(f"Total Salary    : PKR {total_salary}")
print(f"Tax (5%)        : PKR {tax}")
print(f"Net Salary      : PKR {net_salary}")
````

## 📚 Additional Practice Tips

### How to Practice Effectively:

1. **Don't Just Copy Code**: Type everything yourself to build muscle memory
2. **Test with Different Inputs**: Try edge cases (0, negative numbers, very large numbers)
3. **Break Problems Down**: Solve step by step, test each part
4. **Use Comments**: Explain your logic as you code
5. **Make Mistakes**: Errors are learning opportunities
6. **Modify Exercises**: Change requirements to make them harder

### Common Mistakes to Avoid:

- Forgetting to convert `input()` to int/float for calculations
- Using single `=` instead of `==` for comparison
- Not formatting output nicely (use f-strings!)
- Forgetting to test code before submitting
- Not using descriptive variable names
